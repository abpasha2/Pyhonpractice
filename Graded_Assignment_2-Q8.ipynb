```python
# Importing required supporting libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Creating DF object fromn the Application records

df = pd.read_csv('C:\AI-ML\SampleData\Application_record.csv')
```

```python
plt.hist(x='AMT_INCOME_TOTAL',data=df, bins=100)
```

```python
df_mean = np.mean(df.AMT_INCOME_TOTAL)
df_median = np.median(df.AMT_INCOME_TOTAL)
df_std = np.std(df.AMT_INCOME_TOTAL)


print('Mean of df   = {}' .format(df_mean))
print('Median of df = {}' .format(df_median))
print('STD of df    = {}' .format(df_std))
```

```python
df.skew(axis = 0, skipna = True)
```

```python
plt.figure(figsize=(20,8),dpi=200)
# set the histogram, mean and median
sns.distplot(df["AMT_INCOME_TOTAL"], kde=False)
plt.axvline(x=df.AMT_INCOME_TOTAL.mean(), linewidth=3, color='g', label="mean", alpha=0.3)
plt.axvline(x=df.AMT_INCOME_TOTAL.median(), linewidth=3, color='b', label="median", alpha=0.3)

# set title, legends and labels
plt.xlabel("AMT_INCOME_TOTAL")
plt.ylabel("Count")
plt.title("Distribution of Income", size=14)
plt.legend(["mean", "median"])
```
