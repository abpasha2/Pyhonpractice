####  Q9. Create a violin plot of AMT_INCOME_TOTAL by CODE_GENDER?(5 mark)

```python
# Importing required supporting libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Creating df object from the Application records

df = pd.read_csv('C:\AI-ML\SampleData\Application_record.csv')
```

```python
# Drawing violin plot from df dataframe

plt.figure(figsize=(12,6))    
sns.violinplot(x ="CODE_GENDER",
             y ="AMT_INCOME_TOTAL",
             data = df,
              split=True)
```

```python
# Below steps calculate first and third quarter details 

iqr75, iqr25 = np.percentile(df["AMT_INCOME_TOTAL"],[75,25])
iqr=iqr75-iqr25
print('Iqr = {}' .format(iqr))

# Calculate upper and lower limits for outliers
Upper = iqr75 + 1.5*iqr
Lower = iqr25 - 1.5*iqr

print('Upper boundary - {}'.format(Upper))
print('Lower boundary - {}'.format(Lower))

df_iqr = df[(df.AMT_INCOME_TOTAL < Upper) & (df.AMT_INCOME_TOTAL > Lower)]
```

```python
# Drawing violin plot after removing outliers using IQR approach and using 3 variables i.e. AMT_INCOME_TOTAL, NAME_EDUCATION_TYPE, CODE_GENDER 

plt.figure(figsize=(12,6))    
sns.violinplot(y ="AMT_INCOME_TOTAL",
               x="NAME_EDUCATION_TYPE",
             data = df_iqr,
              split=True,
              hue="CODE_GENDER")
```
