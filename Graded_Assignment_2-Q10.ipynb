```python
# Importing required supporting libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Creating DF object fromn the Application records

df = pd.read_csv('C:\AI-ML\SampleData\Application_record.csv')
```

```python
df.describe(include='all',percentiles=[0.01,0.25,0.5,0.75,0.99])
```

```python
df_mean = df.AMT_INCOME_TOTAL.mean()
print('Mean Income value = {}'.format(df_mean))

df_min = df.AMT_INCOME_TOTAL.min()
print('Minimum Income value = {}'.format(df_min))

df_max = df.AMT_INCOME_TOTAL.max()
print('Maximum Income value = {}'.format(df_max))

df_std = df.AMT_INCOME_TOTAL.std()
print('Standard deviation  = {}'.format(df_std))

df_median = df.AMT_INCOME_TOTAL.median()
print('Median value = {}'.format(df_median))


df_q1 = df.AMT_INCOME_TOTAL.quantile(0.25)
print('Q1 value = {}'.format(df_q1))

df_q2 = df.AMT_INCOME_TOTAL.quantile(0.5)
print('Q2 value = {}'.format(df_q2))

df_q3 = df.AMT_INCOME_TOTAL.quantile(0.75)
print('Q3 value = {}'.format(df_q3))
```
