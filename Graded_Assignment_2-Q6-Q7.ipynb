#### Q6. Use the columns AMT_INCOME_TOTAL, NAME_INCOME_TYPE , NAME_EDUCATION_TYPE and infer about the total income based on income type and education type.(5 mark)

```python
# Importing required supporting libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Creating DF object fromn the Application records

df = pd.read_csv('C:\AI-ML\SampleData\Application_record.csv')
```

```python
df.info()
```

```python
plt.figure(figsize=(10,6),dpi=200)
sns.barplot(x='NAME_INCOME_TYPE',y='AMT_INCOME_TOTAL',data=df, hue='NAME_EDUCATION_TYPE', estimator=np.mean,ci='sd')
plt.show()
```

#### Inference from above Graph
#### 1. Commercial associates with Academic degrees draws maximum salary
#### 2. Pensioners with Lower Secondary educations draws minimum pension
#### 3. Commercial associates with any degree/education usually draws more salary(respectively) as compared to other professionals

# Q7. Comment on the salary bias between working FEMALE and MALE.

```python
plt.figure(figsize=(15,6),dpi=200)
sns.barplot(x='NAME_INCOME_TYPE',y='AMT_INCOME_TOTAL',data=df, hue='CODE_GENDER', estimator=np.mean,ci='sd')
plt.show()
```

#### Inference from above graph:
